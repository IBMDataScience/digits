# Import Libraries

```python
import warnings
warnings.filterwarnings("ignore")
%pylab inline
import matplotlib.pyplot as plt

from sklearn.externals import joblib
from sklearn.svm import SVC
from sklearn.datasets import load_digits
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score, train_test_split
```

# Load data 

```python
digits = load_digits()
X = digits.images
X = X.reshape(len(X),-1)
y = digits.target
```

# Visualize images of handwritten digits

```python
fig, axes = plt.subplots(nrows=1, ncols=10, figsize=(20,30))
for i in range(10):
    axes[i].imshow(X[i].reshape(8,8), cmap='gray')
```

# Split data into train and test sets

```python
X_train, X_test, y_train,y_test = train_test_split(X,y)
```

# Training a Classifier

```python
classifier = SVC(gamma=0.001)
classifier.fit(X_train,y_train)
y_predicted = classifier.predict(X_test)
accuracy = round(accuracy_score(y_test, y_predicted)*100,2)
print("Prediction Accuracy :",accuracy,"%")
```

#  5 fold Cross Validation

```python
crossval_accuracies = cross_val_score(classifier, X,y,scoring='accuracy',cv=5)
mean_accuracy = round(mean(crossval_accuracies)*100,2)
print("Mean Accuracy :", mean_accuracy,"%")
```
